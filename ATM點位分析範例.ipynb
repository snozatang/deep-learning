## 寫迴圈將載入的地址資料轉為經緯度，再搜尋鄰近興趣點類別 

```python
atmadds = atmloc_ex['裝設地址']
```

```python
#轉換經緯度
atmxy = []
for atmadd in atmadds:
    loc = gmaps.geocode(atmadd)[0]['geometry']['location']
    lat = gmaps.geocode(atmadd)[0]['geometry']['location']['lat'] 
    lng = gmaps.geocode(atmadd)[0]['geometry']['location']['lng']
    atmxy.append((atmadd, loc, lat, lng))
    
df_xy = pd.DataFrame(atmxy, columns = ['地址', 'loc', 'lat', 'lng'])
```

```python
df_xy.head()
```

```python
import time
```

```python
#發查鄰近興趣點(可以直接用剛得到的loc欄位當key值發查)
#因為API一次只能回傳最多20筆，要使用next_page_token去try拿到完整60筆資料，若已經找不到資料跳keyerror就pass
#如果換頁查詢間沒有留time.sleep會查詢錯誤
atmpoi = []
for row in df_xy.iterrows():
    loc = row[1][1]
    poi_p1 = gmaps.places_nearby(location = loc, radius = 500, language='zh-TW'#,
                                   # type = ['bakery', 'beauty_salon', 'book_store', 'car_rental', 
                                   #         'clothing_store', 'dentist', 'department_store', 'drugstore',
                                   #         'electronics_store', 'furniture_store', 'gas_station', 'gym', 
                                   #         'hardware_store', 'home_goods_store', 'hospital', 'jewelry_store', 
                                   #         'library', 'local_government_office', 'lodging', 'museum', 'park',
                                   #         'restaurant', 'school', 'shoe_store', 'subway_station', 'supermarket',
                                   #         'train_station', 'tourist_attraction']
                                )
    pois = poi_p1['results']
    time.sleep(2)
    try:
        poi_p2 = gmaps.places_nearby(page_token = poi_p1['next_page_token'])
        poi_p2_results = poi_p2['results']
        pois.extend(poi_p2_results)
        time.sleep(2)
        try:
            poi_p3 = gmaps.places_nearby(page_token = poi_p2['next_page_token'])
            poi_p3_results = poi_p3['results']
            pois.extend(poi_p3_results)
        except KeyError:
            pass
    except KeyError:
        pass
    
    poi_list = []
    for poi in pois:
        poi_loc = poi['geometry']['location']
        poi_lng = poi['geometry']['location']['lng']
        poi_lat = poi['geometry']['location']['lat']
        poi_types = poi['types']
        poi_name = poi['name']
        poi_list.append((loc, poi_loc, poi_name, poi_types, poi_lat, poi_lng))                        
    atmpoi.extend(poi_list)

df_poi = pd.DataFrame(atmpoi, columns = ['loc', 'poi_loc', 'poi_name', 'poi_type', 'poi_lat', 'poi_lng'])    
    
```

```python
df_poi.head()
```

```python
df_poi['loc'] = df_poi['loc'].astype(str)
df_poi['poi_loc'] = df_poi['poi_loc'].astype(str)
df_poi['poi_type'] = df_poi['poi_type'].astype(str)
df_xy['loc'] = df_xy['loc'].astype(str)
```

```python
#因google有可能查詢到重覆資料，刪除重複值
df_poi = df_poi.drop_duplicates()
```

```python
df_poi.to_csv('df_poi.csv', encoding = 'Utf-8')
```

## 統計每個地點的興趣點類別 

```python
df_merge = df_xy
```

```python
for spec_type in ['bakery', 'lodging', 'museum', 'park', 'restaurant', 'school', 'supermarket', 'tourist_attraction']:
    poi_cnt = df_poi[df_poi['poi_type'].str.contains(spec_type)].groupby(['loc']).count()
    poi_cnt = poi_cnt.reset_index()[['loc', 'poi_type']]
    poi_cnt = poi_cnt.rename(columns = {'poi_type':spec_type})
    df_merge = pd.merge(df_merge, poi_cnt, left_on = 'loc', right_on = 'loc', how = 'left')
```

```python
df_merge.head()
```

```python
df_merge = df_merge.fillna(0)
```

```python
#merge ATM地點與興趣點
```

```python
df_smy = pd.merge(atmloc_ex, df_merge, left_on = '裝設地址', right_on = '地址', how = 'inner')
df_smy = df_smy[['裝設金融機構名稱', '裝設地點', '裝設地址', 'bakery', 'lodging', 'museum', 'park', 'restaurant', 'school', 'supermarket', 'tourist_attraction']]
```
